```python
def get_WL_data(yr):
    link = 'https://www.basketball-reference.com/leagues/NBA_' + str(yr) + '_standings.html'
    driver = webdriver.Chrome('/Users/danielalpert/Desktop/Thesis/chromedriver')
    driver.get(link)
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    table = soup.find('div', {'id':'all_team_vs_team'})
    divs = table.find_all('div')
    trs = divs[4].find_all('tr')
    wl_results = pd.DataFrame()
    for tr in trs[1:]:
        #print tr
        team = tr.a['href'].split('/')[2]
        print team
        tds = tr.find_all('td')
        for td in tds[1:]:
            #print td
            if td.text != '':
                wl_results = wl_results.append({
                        'team': team,
                        'opp': td['data-stat'],
                        'W': int(td.text.split('-')[0]),
                        'L': int(td.text.split('-')[1])
                    }, ignore_index=True)
                
    season = [yr]*len(wl_results)
    wl_results['yr'] = season
    
    return wl_results
```
